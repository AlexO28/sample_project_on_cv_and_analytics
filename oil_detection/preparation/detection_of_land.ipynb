# Подсчёт статистик при применении эвристики.

Вводим эвристику: третий канал больше 100, либо и первый, и второй каналы меньше 70.

Логика эвристики:

    * третий канал отвечает за синий цвет, он должен быть достаточно насыщенным
    * если первые два канала слабы, то это что-то непонятное, и это лучше считать водой

```python
tab['euristic'] = 0
tab.loc[(tab['mean_B'] >= 100) | (((tab['mean_R'] <= 70) & (tab['mean_G'] <= 70))), 'euristic'] = 1
```

Давайте заимплементим модель детекции суши:
    
    * дробим на мини-области размера 50 на 50
    * считаем средни значения бандов по мини-областям
    * если в мини-области есть нефть, то это мини-область вода
    * если в мини-области 3-й канал, по крайней мере 100, то мини-область вода
    * если в мини-области 1-й и 2-й каналы не более 70, то мини-область вода
    * если снимок чёрно-белый, то мы смотрим на значение единственного канала, он должен быть, по крайней мере, 100
